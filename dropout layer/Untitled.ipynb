```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import confusion_matrix , classification_report
```

```python
df = pd.read_csv("./sonar_dataset.csv", header=None)
df.sample(5)
```

```python
df.info()
```

```python
df[60].value_counts()
```

```python
x=df.drop(60,axis=1)
y=df[60]
```

```python
x.shape
```

```python
y.shape
```

```python
y=y.map({
    'R':0,
    'M':1
})
```

```python
y.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=2)
```

```python
print('shape of x_train',x_train.shape)
print('shape of x_test',x_test.shape)
```

# Without using Dropout layer my model is overfitting the Data

```python
model = keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',
             loss = 'binary_crossentropy',
             metrics=['accuracy'])

model.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred = model.predict(x_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
print(classification_report(y_test, y_pred))
```

# After using the Dropout layer 

```python
model1 = keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid')
])

model1.compile(optimizer='adam',
             loss = 'binary_crossentropy',
             metrics=['accuracy'])

model1.fit(x_train,y_train,epochs=200,batch_size=8)
```

#  After using the Dropout layer the Accuracy and Loss are Improved

```python
model1.evaluate(x_test,y_test)
```

```python
y_pred1 = model1.predict(x_test).reshape(-1)
print(y_pred1[:10])

# round the values to nearest integer ie 0 or 1
y_pred1 = np.round(y_pred1)
print(y_pred1[:10])
```

```python
y_test[:10]
```

```python
print(classification_report(y_test, y_pred1))
```
